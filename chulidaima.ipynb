```python
import pandas as pd
import matplotlib.pyplot as plt

# tab
df = pd.read_csv('dataset_use_1image.txt', sep='\t')
df = df.iloc[1:799] 

# 
plt.figure(figsize=(12, 6))
plt.plot(df['obs_index'], df['inference_time_sec'], marker='o', linestyle='-', markersize=3)
plt.xlabel('')
plt.ylabel('Inference Time (sec)')
plt.title('Inference Time vs obs_index')
plt.grid(True)
plt.show()
```
